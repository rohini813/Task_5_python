```python
#Social Media – Post Engagement Analyzer
```

```python
def analyze(likes_list):
  total=0
  for like in likes_list:
    total+=like
  if total >= 1000:
    status="Viral Post"
  else:
    status="Normal Engagement"
  print("Total Likes:",total)
  print("Post Status:",status)
likes=[200, 300, 150, 400]
analyze(likes)
```

```python
#Agriculture – Rainfall Adequacy Checker
```

```python
def rainfall_checker(rainfall_list, required):
    total=0

    for r in rainfall_list:
        total+=r

    avg=total/len(rainfall_list)

    if avg>= required:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainfall"

    print("Average Rainfall:",avg)
    print("Rainfall Status:",status)

rainfall=[60, 75, 80, 70, 75]
rainfall_checker(rainfall, 70)
```

```python
#Social Media – Duplicate Account Detection
```

```python
def check_duplicates(usernames):
    unique_users=set(usernames)

    if len(unique_users)!=len(usernames):
        result= "Yes"
    else:
      result="No"

    print("Duplicate Account Found:",result)

user=["rohan", "amit", "rohan", "sneha"]
check_duplicates(user)
```

```python
# System Monitoring – Application Health Checker
```

```python
def system_health(errors):
   if errors==0:
      status="Healthy"
   elif errors<=5:
     status="Minor Issues"
   else:
    status="Critical Issues"

   print("Error count:",errors)
   print("System status:",status)

errors=int(input("Enter the number of errors:"))
system_health(errors)
```

```python
# E-Learning – Student Attendance Eligibility System
```

```python
def attendance_check(attendance_list):
  attended_classes=len(attendance_list)
  attended=0
  for status in attendance_list:
    if status=="Present":
      attended+=1

  precentage=(attended/attended_classes)*100

  if precentage>=75:
    status="Eligible"
  else:
    status="Not Eligible"

  print("Attendance Precentage:",precentage)
  print("Eligibility Status:",status)

attendance=["Present", "Present", "Absent", "Present"]
attendance_check(attendance)
```
